```python
import sqlite3
import pandas as pd
```

```python
connection = sqlite3.connect("mental_health.sqlite")
```

```python
query = '''SELECT * from Answer as ans
left join Question as ques
on ans.QuestionID = ques.questionid
where ans.SurveyID = 2014
and ans.QuestionID in (1,2,3,5,6,7,13,20,21)'''
```

```python
survey = pd.read_sql_query(query, connection)
_, _1 = survey.pop('SurveyID'), survey.pop('questionid')  # We don't need these data
schema = ['UserID', 'QuestionID', 'questiontext', 'AnswerText' ]
survey = survey[schema]
user_ids = []
for x in survey['UserID']:
    if x not in user_ids:
        user_ids.append(x)
```

```python
survey[(survey['UserID'] == 1)]  # A sample on which below code is working
```

```python
users = {}
for i in user_ids:
    data = survey[(survey['UserID'] == i)]
    _ = data.pop('UserID')
    temp_data = []
    for j in data.values:
        temp_data.append(list(j))
    users[i] = temp_data
```
